```python
import ipywidgets as widgets
from IPython.display import display
import pandas as pd
```

```python
scores = {}
# Add player
new_player = widgets.Text(
    description='Player:',
    placeholder='Enter name'
)

add_player_btn = widgets.Button(
    description='Add Player',
    button_style='success'
)

# Select player + points
player_dropdown = widgets.Dropdown(
    description='Player:',
    options=[]
)

points_input = widgets.IntText(
    description='Points:',
    value=1
)

add_points_btn = widgets.Button(
    description='Add Points',
    button_style='primary'
)

# Output area
score_output = widgets.Output()
def update_scoreboard():
    score_output.clear_output()
    with score_output:
        if scores:
            df_scores = (
                pd.DataFrame.from_dict(scores, orient='index', columns=['Score'])
                .sort_values('Score', ascending=False)
            )
            display(df_scores)
        else:
            print("No players yet.")
def add_player_clicked(b):
    name = new_player.value.strip()
    if name and name not in scores:
        scores[name] = 0
        player_dropdown.options = list(scores.keys())
        new_player.value = ''
        update_scoreboard()

def add_points_clicked(b):
    player = player_dropdown.value
    if player:
        scores[player] += points_input.value
        update_scoreboard()

add_player_btn.on_click(add_player_clicked)
add_points_btn.on_click(add_points_clicked)
```

```python
display(
    widgets.VBox([
        widgets.HBox([new_player, add_player_btn]),
        widgets.HBox([player_dropdown, points_input, add_points_btn]),
        score_output
    ])
)

update_scoreboard()
```
